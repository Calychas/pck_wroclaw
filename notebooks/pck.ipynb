```python
df_pck = gpd.read_parquet(data_dir / "pck.parquet")
df_pck
```

```python
df_pck_500 = gpd.read_parquet(data_dir / "pck_500.parquet")
df_pck_500
```

```python
m = KeplerGl(
    height=800,
    data={
        "PCK": df_pck.drop(columns="address").copy(),
        "PCK_500": df_pck_500.drop(columns="address").copy(),
    },
)
m
```

```python
m.save_to_html(
    data={
        "PCK": gpd.GeoDataFrame(data={"address": df_pck.address.to_list()}, geometry=df_pck.geometry, crs="epsg:4326"),
        "PCK_500": gpd.GeoDataFrame(data={"address": df_pck_500.address.to_list()}, geometry=df_pck_500.geometry, crs="epsg:4326"),
    },
    file_name=data_dir.parent / "pck.html",
    center_map=True,
)
```
